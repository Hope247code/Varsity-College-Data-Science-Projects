# What is Classification:

### Definition of Classification in Machine Learning:
Classification modelling is a machine learning method that uses supervised learning methods, such as pattern recognition and categorization, on labaled data to try and predict probability that the output will have a certain label given that similar input data is given to the model (Keita 2022) and (Banoula 2023). 

Classification Algorithms as explained by Banoula (2023) is part of the supervsed learning methods used in the training of classification models, these algorithms helps us to categorize the data into usable groups as well as detecting patterns that is present in the data, that will help us group the observed values into groups that is similar, this will be used to create a prediction model that will give us the probability of a label, given similar input data.

**### Why Classification?**

The classification models help us to categorize and organize data in a simplified structured manner, to gain insight from the factors that influences the probability of the observations being placed in a certain category (Ramakrishnan 2023). 


**### Classification Modelling Concepts:**

**1. Supervised Learning Approach**

Supervised Learning in machine learning is done by training a model with input and given output, thus it is trained on examples on what the answer/output should be, to be able to predict the output after training with similar inputs (Banoula 2023).

Examples:
- Regression
- Classification

**2. Unsupervised Learning**

Unsupervised Learning is used by training a model by giving the model all that data, so that the model can look for insights and patterns from the data no its own.

Examples:
- Clustering
- Dimensionality reduction

**3. Semi-supervised Learning**

Semi-supervised Learning is between the unsupervised and supervised, it is a combination of the two that gives us uniques algorithms to work with.

**4. Reinforcement Learning**

Reinforcement Learning is when we create an aloralgorithm that improves on itself, by frequently repeating the training on itself and looking at the output accuracy. This model has no fixed input and can improve as the input increases or decreases over time without breaking.

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score

# model evaluation calculate and score
from sklearn.metrics import accuracy_score, recall_score, precision_score, roc_auc_score,  mean_squared_error
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import svm

# plot metrics model answer(metrics)
from sklearn.metrics import plot_roc_curve, plot_precision_recall_curve

# machine learning model
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier

```

```python
# Define X and y
X = df.drop(['isFraud'], axis=1)
y = df['isFraud']
```

We are going to split the data where 30% will be used as testing data, to test our predictions 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

One can check the shapes of train and test sets with the following code:

```python
print("x_train shape: ", X_train.shape )
print("x_test shape: ", X_test.shape )
print("y_train shape: ", y_train.shape )
print("y_test shape: ", y_test.shape )
```

### 5.2 SMOTE and Scaling to Balance data

```python
smote = SMOTE(sampling_strategy='minority', random_state=42)
X_train_re, y_train_re = smote.fit_resample(X_train, y_train)
```

```python
sns.histplot(y_train_re)
```

As can be seen in the graph above this dataset has now been balanced and can be used to create an classification model using the Naive Bayes classifier

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Create a scaler object
scaler = StandardScaler()

# Fit the scaler on the training data
scaler.fit(X_train_re)

# Transform the training and testing data
X_train_scaled = scaler.transform(X_train_re)
X_test_scaled = scaler.transform(X_test)
```

### 5.3 Train the Naive Bayes model

1. Results and Interpretation of the model
2. accuracy, precision, recall, and F1-score
3. Can we improve the model?
4. Conclusion

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_NB = GaussianNB()
model_NB.fit(X_train_scaled, y_train_re)
y_pred_NB = model_NB.predict(X_test)
```

```python
# Quick evaluation
Train_Accuracy = model_NB.score(X_train_scaled, y_train_re)
Test_Accuracy = model_NB.score(X_test_scaled, y_test)
print(f'Training accuracy: {Train_Accuracy*100:.2f} %')
print(f'Testing accuracy: {Test_Accuracy*100:.2f} %')
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred_NB).sum()))
```

```python
gnd_accuracy = accuracy_score(y_test,y_pred_NB)
print(gnd_accuracy)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, f1_score
```

```python
# Confusion Matrix
CM = confusion_matrix(y_true=y_test, y_pred=y_pred_NB)
ConfusionMatrixDisplay(CM, display_labels=data['isFraud'].unique()).plot()
plt.title('Confusion Matrix Without Normalization')
plt.show()
```

```python
gnd_report = classification_report(y_test,y_pred_NB)
print(gnd_report)
```

```python

# Accuracy = (TP + TN) / (TP + TN + FP + FN)
Accuracy_NB = accuracy_score(y_test, y_pred_NB)
print(f'âž¤âž¤âž¤ Accuracy Score : {Accuracy_NB * 100 : .2f} %\n')


# Precision = TP / (TP + FP)
Precision_NB = precision_score(y_test, y_pred_NB)
print(f'âž¤âž¤âž¤ Precision Score : {Precision_NB * 100 : .2f} %')
print('Accuracy of fraud detected\n')


# Recall = TP / (TP + FN)
Recall_NB = recall_score(y_test, y_pred_NB)
print(f'âž¤âž¤âž¤ Recall Score : {Recall_NB * 100 : .2f} %')
print('the cases of fraud matched\n')

# F1 Score = 2 Ã— ((Precision * Recall) / (Precision + Recall))
F1_Score_NB = f1_score(y_test, y_pred_NB)
print(f'âž¤âž¤âž¤ F1 Score : {F1_Score_NB * 100 : .2f} %')
print('fraud predictions were correct\n')

ROC_AUC_NB = roc_auc_score(y_test, y_pred_NB)
print(f'âž¤âž¤âž¤ AUC_ROC : {ROC_AUC_NB * 100 : .2f} %\n')
```

```python
# metrics method
def metrics_calculate(model_name, y_val, y_pred):
    '''
    0. basic metrics values ['accuracy', 'precision', 'recall', 'fpr', 'fnr', 'auc']
    1. classification report
    2. confusion matrix
    '''
    tn, fp, fn, tp = confusion_matrix(y_val, y_pred).ravel()
    fpr = fp / (tn + fp)
    fnr = fn / (tp + fn)
    
    print(model_name, 'classification report:\n', classification_report(y_val, y_pred))
    print(model_name, 'confusion_matrix:\n', confusion_matrix(y_val, y_pred))
    print('\n%s FNR:%.8f FPR:%.8f\n%s accuracy:%.8f' % (model_name, fnr, fpr, model_name, accuracy_score(y_pred,y_val)))
```

```python
kfold = KFold(n_splits=5, random_state=42, shuffle=True)
# get model -- GaussianNB
gaussion_nb = GaussianNB()
cv = cross_val_score(gaussion_nb, X_train_re, y_train_re, cv=kfold, scoring='accuracy')
print('Naive Bayes GaussianNB cv score: ', cv)
```

```python
# fit
gaussion_nb.fit(X_train_re, y_train_re)
# predict
y_pred = gaussion_nb.predict(X_test)
# evaluate
metrics_calculate('Naive Bayes - GaussionNB', y_test, y_pred)
```
