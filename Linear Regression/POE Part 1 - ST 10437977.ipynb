# What is Linear Regression:

#### Linear Regression Definition:
James, Witten, Hastie, Tibshirani and Taylor (2023) explains that Linear Regression is a statistical learning model that is still used in a lot of different fields, this method is the simplest method to use to do supervised learning if given quantitative data, linear regression is also used with quantitative data to do predictive modelling.
Linear regression uses quantitative data to determine if the dependent variable and the independent variable has a correlation if the independent variable has an impact on the dependent variable. We use the quantitative data to create a model to predict the next quantitative data point of the dependent variable, based on the independent variable. 

The simplest equation for this model is (What is linear regression? | IBM n.d.):
        
        Y ≈ β0 + β1X1 + β2X2 + …+ βpXp

X : is the independent variable(s) 
Y : is the dependent variable = (the predicted score)
B0 : is the intercept = (where the line will go through the y-axis at value x=0)
βi : is the regression coefficients estimates of each independent variable (the slope)

In (What is linear regression? | IBM n.d.) they explain that we use linear regression to analyse what the coefficients of this model may be, that can be used together with the independent variable(s) to predict the next dependent variable's response.
Linear regression analysis is used to see what the impact the different independent variables have on the dependent variable and each other, this is represented in the coefficient of the independent variable, the dependent variable will change according to the independent variable based on the coefficient value.  Kumari and Yadav (2018) explain to us that we use the linear regression model to analyse the associations between each independent variable and the dependent variable to create predictor estimates. ## Linear Regression Concepts
(Biswal 2023)

1. **Best Fit** : the straight line in a plot that minimizes the divergence between related dispersed data points
2. **Coefficient** : A coefficient in linear regression represents changes in a Response Variable.
3.  **Coefficient of Determination** : The correlation coefficient. The precision or degree of fit.
4. **Correlation** : Measures the intensity and degree of association between two variables, range from -1.0 to 1.0.
5. **Dependent Feature** : A variable represented as y in the slope equation y=ax+b. Also referred to as an Output or a Response.
6. **Estimated Regression Line** : The straight line that best fits a set of randomly distributed data points.
7. **ndependent Feature** : a variable represented by the letter x in the slope equation y=ax+b. Also referred to as an Input or a predictor.
8. **Intercept** : the point at where the slope intersects the Y-axis, indicated by the letter b in the slope equation y=ax+b.
9. **Least Squares** : A method for calculating the best fit to data by minimizing the sum of the squares of the discrepancies between observed and estimated values.
10. **Mean** : an average of a group of numbers, Mean is represented by a linear function.
11. **OLS (Ordinary Least Squares Regression)** :  known as Linear Regression.
12. **Residual** : the vertical distance between a data point and the regression line.
13. **Regression** : is an assessment of a variable's predicted change in relation to changes in other variables.
14. **Regression Model** : The optimum formula for approximating a regression.
15.  **Response Variables** : This category covers both the Predicted Response (the value predicted by the regression) and the Actual Response (the actual value of the data point) 
16. **Slope** : the steepness of a regression line. The linear relationship between two variables may be defined using slope and intercept: y=ax+b
17. **Simple linear regression** : A linear regression with a single independent variable.


## Evaluating the Model

1. Mean Squared Error (MSE): This is the most commonly used metric for evaluating a Linear Regression model. It measures the average of the squared differences between the predicted values and the actual values. A lower MSE indicates better performance.

2. Root Mean Squared Error (RMSE): This is the square root of the MSE and provides a more interpretable metric since it is in the same units as the target variable.

3. R-squared (R2): This metric measures the proportion of variance in the target variable explained by the model. An R2 score of 1 indicates a perfect fit, while a score of 0 indicates that the model is no better than predicting the mean value of the target variable.

4. Mean Absolute Error (MAE): This metric measures the average absolute differences between the predicted and actual values. It is less sensitive to outliers than the MSE.

##### 5.1.1 Cross-Validation and Hyperparameter Tuning

This is a method used to reduce the risk of overfitting or underfitting the data. This is done by partitioning the data into k(amount) different sets.

- A model is trained using (k-1) of the folds as training data
- the resulting model is validated on the remaining part of the data (to compute a performance measures)
-  Cross-validation provides a more comprehensive understanding of a model’s performance across different subsets of the data.
- helps assess a model’s robustness
- a balance between bias and variance

**r2_score** : The proportion of the variance in the dependent variable (charges)  that is predictable from the independent variable(s)

```python
# Import the metrics libraries that we will use to eveluate our model
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

- Mean Squared Error (MSE): This is the most commonly used metric for evaluating a Linear Regression model. It measures the average of the squared differences between the predicted values and the actual values. A lower MSE indicates better performance.

- Root Mean Squared Error (RMSE): This is the square root of the MSE and provides a more interpretable metric since it is in the same units as the target variable.
R-squared (R2): This metric measures the proportion of variance in the target variable explained by the model. An R2 score of 1 indicates a perfect fit, while a score of 0 indicates that the model is no better than predicting the mean value of the target variable.

- Mean Absolute Error (MAE): This metric measures the average absolute differences between the predicted and actual values. It is less sensitive to outliers than the MSE.
