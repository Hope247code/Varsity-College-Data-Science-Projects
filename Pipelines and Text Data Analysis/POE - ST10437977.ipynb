**Step 1 : Data Preprocessing**

* Data Cleaning     
    
    - Removing noise: This is done by removing punctuation marks, tags and characters that is not readable by machine learning models such as emojies and HTML format indicators that is used when publishing on websites (Importance of text pre-processing n.d.).

    - Removing stopwords: This is done to get the most relevant words needed for the training and analysis of text data (Nabi 2019). 

        - *Stopwords* : This is a list made up out of the most common used words in the English language such as : 'and' , 'is', 'the', these words do not have a spesific sematic importance (Nabi 2019). 

    - Lowercase all text: This is done to simplify the text data for machine learninng models to understand and interpret the data better. Machine learning models will classify "Hello" and "hello" as two different words, this is because capitalized words are read differently by machines (Importance of text pre-processing n.d.). 

* Stemming 
    - This is a method done to remove additional characters such as a suffix from a word (Nabi 2019).

* Lemmatization     
     - This is a method of finding the root meaning of each word to find the most simplified part of the word for easier text processing and analysis (Nabi 2019).
     - Sematic information gives us the meaning of a word, which helps us and the machine leanring model to know what the meaning of a piece of text are (Importance of text pre-processing n.d.).

**Step 2: Feature Extraction**

* Bag of Words (BOW)
    -  This method is used to extract features out of the text data. This is done to represent the text numerically by creating a list of unique words found in the text and counting them to create a vocabulary for all the reviews that was collected (Nabi 2019).
    - The Term Frequency-Inverse Document Frequency (TF-IDF) technique is the most popular (Nabi 2019). Here is the mathematical representation:
        - Term Frequency (TF) = (Number of times term t appears in a document)/(Number of terms in the document)
        - Inverse Document Frequency (IDF) = log(N(number of documents)/n(number of documents a term t has appeared in))
        - TF-IDF value of *a term* = TF * IDF

**What is pipelines in Machine Learning?**

A pipeline is a method for organizing a sequence of operations or functions that process data. Each operation's output serves as the input for the next, continuing in this manner until the final result is produced. You can visualize a pipeline as a chain of pipes, where data flows from one pipe to the next, undergoing various transformations or manipulations along the journey (Thakur 2023).

**Why use Pipelines?**

Readability: Pipelines make the code more readable by clearly showing the flow of data from one operation to the next, without cluttering it with intermediate variables or loops (Thakur 2023) .

Efficiency: Pipelines can improve code efficiency by avoiding the creation of intermediate lists or iterables, which take up memory and slow down execution. Operations are executed only when needed, saving time and resources, especially with large or infinite data sources (Thakur 2023).

Flexibility: Pipelines are easily modified or extended, allowing for the addition, removal, or reordering of operations without affecting the rest of the code. They support various operations like filtering, mapping, reducing, aggregating, grouping, and sorting, which can be combined in different ways to achieve various results (Thakur 2023).

## Step 3 : Data Preprocessing
* Tokenization — convert sentences to words
* Removing unnecessary punctuation and tags
* Removing stop words 
* Stemming - reduce word to its root by dropping unnecessary characters, such a suffix
* Lemmatization — removing inflection which involves identifying the part of speech and using a comprehensive language database.

Normally any NLP task involves following text cleaning techniques -
1. Removal of punctutions and special characters
2. Removal of stopwords 
3. change all text to lowercase letters
4. Lemmatization 
5. Tokenization
6. Bag of words vectorization - Encode the numeric values once you have cleaned it.


```python
def clean_text(doc):
	# split into tokens by white space
	tokens = re.sub("[^\w]", " ", doc).split()
	# remove punctuation from each token
	table = str.maketrans('', '', string.punctuation)
	tokens = [w.translate(table) for w in tokens]
	# remove remaining tokens that are not alphabetic
	tokens = [word for word in tokens if word.isalpha()]
	#  lowercase letters
	tokens = [word.lower() for word in tokens]
	# filter out stop words
	stop_words = set(stopwords.words('english'))
	tokens = [w for w in tokens if not w in stop_words]
	# Stemming   (GeeksforGeeks 2023)
	porter = PorterStemmer()
	tokens = [porter.stem(word) for word in tokens]
	# Lemmatize tokens
	wordnet_lemmatizer = WordNetLemmatizer()
	tokens = [wordnet_lemmatizer.lemmatize(word) for word in tokens]	
	# filter out short tokens
	tokens = [word for word in tokens if len(word) > 1]
	# convert back to a string
	tokens = " ".join(tokens)
	return tokens
```

### Term Frequency-Inverse Document Frequency model (TFIDF)

It is used to convert text documents to matrix of tfidf features.

(Dave 2023)

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer,TfidfTransformer
```

```python
cvec = CountVectorizer(stop_words='english', min_df=1, max_df=.5, ngram_range=(1,2))
cvec
```

```python
# Calculate all the n-grams found 
from itertools import islice
cvec.fit(data['review_clean'])
list(islice(cvec.vocabulary_.items(), 20))
```

```python
# Check how many total n-grams we have
len(cvec.vocabulary_)
```

```python
# transform the document into a “bag of words”
cvec_counts = cvec.transform(data['review_clean'])
print( 'sparse matrix shape:', cvec_counts.shape)
print( 'nonzero count:', cvec_counts.nnz)
print('sparsity: %.2f%%' % (100.0 * cvec_counts.nnz / (cvec_counts.shape[0] * cvec_counts.shape[1])))
```

The more sparse the data is the more challenging it will be to model

```python
# Let’s look at the top 20 most common terms

occ = np.asarray(cvec_counts.sum(axis=0)).ravel().tolist()
counts_df = pd.DataFrame({'term': cvec.get_feature_names(), 'occurrences': occ})
counts_df.sort_values(by='occurrences', ascending=False).head(20)
```

```python
counts_df.to_csv("top 20 most common terms.csv")
```

```python
# use the TfidfTransformer to calculate the weights for each term
transformer = TfidfTransformer()
transformed_weights = transformer.fit_transform(cvec_counts)
transformed_weights
```

```python
# And we can take a look at the top 20 terms by average tf-idf weight

weights = np.asarray(transformed_weights.mean(axis=0)).ravel().tolist()
weights_df = pd.DataFrame({'term': cvec.get_feature_names(), 'weight': weights})
weights_df.sort_values(by='weight', ascending=False).head(20)
```

```python
weights_df.to_csv("top 20 terms by average tf-idf weight.csv")
```

```python
#Predict the class labels for the test data using the Multinomial Naive Bayes model
y_pred_lg =pipeline_lg.predict(X_test)

#Calculates the accuracy of the predicted labels compared to the true labels
# Accuracy = (TP + TN) / (TP + TN + FP + FN)
accuracy_lg = accuracy_score(y_test,y_pred_lg)
print(f'➤ Accuracy Score : {accuracy_lg * 100 : .2f} %\n')

#Calculate the precision of the predicted labels compared to the true labels
# Precision = TP / (TP + FP)
precision_lg = precision_score(y_test,y_pred_lg)
print(f'➤ Precision Score : {precision_lg * 100 : .2f} %\n')

#Calculates the recal of the predicted labels compared to the true labels
# Recall = TP / (TP + FN)
recall_lg = recall_score(y_test,y_pred_lg)
print(f'➤ Recall Score : {recall_lg * 100 : .2f} %\n')


# F1 Score = 2 × ((Precision * Recall) / (Precision + Recall))
F1_Score_lg = f1_score(y_test, y_pred_lg)
print(f'➤ F1 Score : {F1_Score_lg * 100 : .2f} %\n')


ROC_AUC_lg = roc_auc_score(y_test, y_pred_lg)
print(f'➤ AUC_ROC : {ROC_AUC_lg * 100 : .2f} %\n')

report_lg = classification_report(y_test, y_pred_lg)
print("classification report:")
print(report_lg)
```
