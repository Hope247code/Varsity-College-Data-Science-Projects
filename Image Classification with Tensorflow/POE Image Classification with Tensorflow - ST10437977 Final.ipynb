## **RICE Image Classification**

#### **Explanation of what image classification is:**

Image classification is a core task in computer vision that involves assigning a specific category or label to an entire image based on its visual content. The process starts with analyzing the image and sorting it into predefined categories, like “cats,” “dogs,” or “mountains,” using computer algorithms. These algorithms go beyond recognizing basic shapes and colors; they extract features such as textures, edges, and patterns to identify the depicted objects or scenes. To train the model, vast datasets of labeled images are used, helping the algorithm learn the distinguishing characteristics of various categories. (GeeksforGeeks, 2024)

**Types of Image Classification:**

(GeeksforGeeks, 2024)

There are several types of image classification depending on the complexity of the task. 
- **Binary classification** is the simplest form, where images are classified into one of two categories, such as determining whether an image contains a cat or not. 
- **Multiclass classification** involves categorizing images into more than two classes, such as identifying different types of animals. 
- **Multilabel classification** allows an image to have multiple labels, like labeling an image as both “sunset” and “beach.” 
- **Hierarchical classification** organizes images into multiple levels, such as classifying an animal first as a "mammal" and then as a "dog." 
- **Fine-grained classification** focuses on distinguishing between similar categories, like identifying specific breeds of dogs. 
- **Zero-shot classification** allows the model to classify images into categories it has never seen before, using semantic relationships between known categories. 
- **Few-shot classification** enables the model to classify images with only a few examples per category, which is useful in cases where labeled data is limited.

#### **Justification for Using Transfer Learning in Rice Dataset Classification**
(Cheng, 2022)
1. Limited Dataset Size and Complexity: 
    - Transfer learning leverages pre-trained models, which have already learned useful features from large datasets, helping improve performance when the rice dataset alone might not be sufficient to train a deep model from scratch.
2. Faster Training and Reduced Overfitting: 
    - Pre-trained models reduce training time and computational costs by starting with learned features. Fine-tuning these models on the rice dataset helps prevent overfitting, especially with a relatively smaller dataset.
3. Better Generalization: 
    - Transfer learning improves generalization by using features learned from diverse data, allowing the model to recognize general visual patterns that are helpful in classifying rice varieties despite the dataset’s specific focus.
4. Limited Need for Extensive Data Augmentation: 
    - Since pre-trained models already extract robust features, transfer learning requires less aggressive data augmentation, making the model more efficient.
5. Proven Success in Image Classification: 
     -Transfer learning has consistently delivered strong results in image classification tasks across various domains, making it a reliable choice for the rice dataset.
6. Focus on Rice-Specific Features: 
    - Fine-tuning higher layers of pre-trained models allows the model to focus on rice-specific features, like texture and color, while still benefiting from the powerful feature extraction learned from large-scale datasets.

Using transfer learning for the rice classification dataset allows for faster model development, improved performance with fewer resources, and better generalization. By leveraging pre-trained networks and fine-tuning them to classify rice varieties, we can achieve high accuracy and robustness in distinguishing between different rice types, making it a powerful strategy for this problem.

```python
# Set the path to the dataset
dataset_path = 'Rice_Image_Dataset'

# List of rice classes
rice_classes = ['Arborio', 'Basmati', 'Ipsala', 'Jasmine', 'Karacadag']

# Create a list of tuples that contain image paths and their class labels
image_data = []  # Initialize an empty list for storing image paths and labels

# Iterate through each class in the detection
for class_name in rice_classes:
    class_path = os.path.join(dataset_path, class_name)  # Create the path for the class
    files = os.listdir(class_path)  # List all files in the class folder
    
    # Loop through all files and add their paths and class labels to the image_data list
    for file in files:
        file_path = os.path.join(class_path, file)  # Create the full file path
        image_data.append((file_path, class_name))  # Append tuple to the list

# Convert the list to a pandas DataFrame
df = pd.DataFrame(image_data, columns=['image_paths', 'label'])

# Shuffle the DataFrame rows for randomness
df = df.sample(frac=1).reset_index(drop=True)
```

```python
# https://www.analyticsvidhya.com/blog/2022/04/master-image-augmentation-with-widely-used-python-libraries/
#  author: Devashree
import imageio
import imgaug as ia
import imgaug.augmenters as iaa
import ipyplot

input_img = imageio.imread(r"Rice_Image_Dataset\Basmati\basmati (4).jpg")

#Horizontal Flip
hflip= iaa.Fliplr(p=1.0)
input_hf= hflip.augment_image(input_img)
#Vertical Flip
vflip= iaa.Flipud(p=1.0) 
input_vf= vflip.augment_image(input_img)
# Image Rotation
rot1 = iaa.Affine(rotate=(-50,20))
input_rot1 = rot1.augment_image(input_img)
# Crop
crop1 = iaa.Crop(percent=(0, 0.3)) 
input_crop1 = crop1.augment_image(input_img)
# Adding Noise to Images
noise=iaa.AdditiveGaussianNoise(10,40)
input_noise=noise.augment_image(input_img)

shear = iaa.Affine(shear=(-40,40))
input_shear=shear.augment_image(input_img)

contrast=iaa.GammaContrast((0.5, 2.0))
contrast_sig = iaa.SigmoidContrast(gain=(5, 10), cutoff=(0.4, 0.6))
contrast_lin = iaa.LinearContrast((0.6, 0.4))
input_contrast = contrast.augment_image(input_img)

sigmoid_contrast = contrast_sig.augment_image(input_img)
linear_contrast = contrast_lin.augment_image(input_img)

images_list=[input_img, input_hf, input_vf,input_rot1,input_crop1,input_noise, input_contrast,sigmoid_contrast,linear_contrast, input_shear]
labels = ['Original', 'Horizontally flipped', 'Vertically flipped','Rotated Image', 'Cropped Image', 'Gaussian Noise Image', 'Gamma Contrast','SigmoidContrast','LinearContrast', 'Image Shearing']
ipyplot.plot_images(images_list,labels=labels,img_width=180)
```

## Step 5: Modelling

### Step 5.1: Build a Classic CNN Image Classsification model

##### Convolutional Neural Network (CNN) model using TensorFlow
(Image Classification, n.d.)
##### **Model architecture**

**Convolutional Layers (Conv2D)**
- layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(224, 224, 3))
    - 32: The number of filters used in the convolutional layer. Each filter detects a specific feature in the input image.
    - kernel_size=(3, 3): The size of the filter (3x3 pixels). Smaller filters (like 3x3) are common as they focus on small details.
    - activation='relu': The activation function applied to the output of the convolution operation. ReLU introduces non-linearity and helps the model learn complex patterns.
    - input_shape=(224, 224, 3): Specifies the input image dimensions: 224x224 pixels with 3 color channels (RGB).
- Second Conv2D Layer
    - Increases the number of filters to 64 to learn more detailed and complex features from the image. The kernel_size and activation remain the same.

**Pooling Layers (MaxPooling2D)**
- layers.MaxPooling2D(pool_size=(2, 2))
    - pool_size=(2, 2): The pooling window size. It reduces the spatial dimensions of the feature maps by taking the maximum value in each 2x2 block, effectively downsampling the data.
    - Purpose: Pooling reduces the computational load, minimizes overfitting, and retains the most important features.
- MaxPooling layers follow each convolutional layer to reduce the dimensions of the feature maps while maintaining important spatial information.

**Flatten Layer (Flatten)**
- layers.Flatten()
    - Converts the 2D feature maps from the convolutional layers into a 1D vector. This is necessary to connect the convolutional layers to the dense layers for classification.

**Dense Layers (Dense)**
- layers.Dense(128, activation='relu')
    - 128: The number of neurons in the layer. This is a fully connected layer that learns high-level features from the flattened input.
    - activation='relu': The ReLU activation function is used for non-linear transformations.
- layers.Dense(5, activation='softmax')
    - 5: The number of neurons matches the number of output classes (e.g., 5 image categories).
    - activation='softmax': Converts the output into probabilities, making it suitable for multi-class classification.

##### **Model Compilation**
- loss='categorical_crossentropy'
    - Loss function used for multi-class classification problems where the labels are one-hot encoded.
- optimizer='adam'
    - Adam optimizer is used for training. It combines the benefits of RMSprop and SGD optimizers, making it well-suited for most scenarios.
- metrics=['accuracy']
    - Monitors the accuracy during training and validation to evaluate model performance.



```python
# attribute: https://medium.com/@sehjadkhoja0/title-exploring-and-analyzing-image-data-with-python-79a7f72f4d2b
# aurthor: sehjadkhoja0
# error handling from https://stackoverflow.com/questions/74792455/module-keras-layers-has-no-attribute-experimental

from tensorflow.keras import layers, models

# Create the model
model = models.Sequential([
    layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(224, 224, 3)),    
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(5, activation='softmax')
])

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Step 6.0: Build Pipeline for image augmentation and preprocessing



```python
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
import cv2

class ResizeTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, img_size):
        self.img_size = img_size

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Resize images
        return np.array([cv2.resize(img, self.img_size) for img in X])

class RandomCropTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, crop_size):
        self.crop_size = crop_size

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Random crop
        cropped_images = []
        for img in X:
            h, w, _ = img.shape
            top = np.random.randint(0, h - self.crop_size[0] + 1)
            left = np.random.randint(0, w - self.crop_size[1] + 1)
            cropped_images.append(img[top:top + self.crop_size[0], left:left + self.crop_size[1]])
        return np.array(cropped_images)

class RandomFlipTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, horizontal=True, vertical=False):
        self.horizontal = horizontal
        self.vertical = vertical

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        flipped_images = []
        for img in X:
            if self.horizontal and np.random.rand() > 0.5:
                img = cv2.flip(img, 1)  # Horizontal flip
            if self.vertical and np.random.rand() > 0.5:
                img = cv2.flip(img, 0)  # Vertical flip
            flipped_images.append(img)
        return np.array(flipped_images)

class ToTensorTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Convert to tensor-like format (H, W, C) to (C, H, W)
        return np.array([img.transpose(2, 0, 1) for img in X])


from sklearn.pipeline import Pipeline

IMG_SIZE = (224, 224)
CROP = (224, 224)

train_pipeline = Pipeline([
    ('resize', ResizeTransformer(img_size=IMG_SIZE)),
    ('random_crop', RandomCropTransformer(crop_size=CROP)),
    ('random_flip', RandomFlipTransformer(horizontal=True, vertical=True)),
    ('to_tensor', ToTensorTransformer()),
])

test_val_pipeline = Pipeline([
    ('resize', ResizeTransformer(img_size=IMG_SIZE)),
    ('center_crop', RandomCropTransformer(crop_size=CROP)),  # Center cropping can be implemented separately if needed
    ('to_tensor', ToTensorTransformer()),
])

# Example image data
X_train = [np.random.rand(256, 256, 3) * 255 for _ in range(10)]  # Dummy data
X_train = np.array(X_train, dtype=np.uint8)

# Apply the train pipeline
processed_train_images = train_pipeline.fit_transform(X_train)

# Apply the test/validation pipeline
processed_test_images = test_val_pipeline.transform(X_train)

```

- Overfitting: A small gap between training and validation loss might suggest that the model is not generalizing well.
