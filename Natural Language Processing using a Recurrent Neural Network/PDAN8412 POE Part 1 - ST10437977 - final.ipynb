# Recurent Neural Network:

#### **What is a Recurent Neural Network?**

Recurrent Neural Network models is a deep neural network that has been trained on sequential data to create a machine leanrning model to make predictions based on input. Recurrent Neural Networks are identified by the fact that it has "memory", thus it can retain what it has learned from the training data. The recurrent neural networks sequential training that depends on all the prior nodes and layers that is dependend on each other. The main two elemants that gives problems to the model are the exploding and vanishing gradients that influence the rate in which the model is learning, te vanishing gradient becomes smaller as the model learns and as the gradient dereases the rate of leanring decreases untill it stops learning, the exploding gradient is the opisite the gradient increases until it is to large and creates an model that is unstable and not reliable, this can be fixed by decreasing the number of layers in the RNN (What is a recurrent neural network (RNN)? | IBM, 2024).  

How RNN's Work:
1. Sequential Processing is done on data by looking at the inputs in the sequence that they are given and the layers in which they are "recognized" to determine the output of that input (Durna, 2024).
2. The RNN looks for recurrent connections that forms patterns and is stored in the models "memory" that is gained by the patterns learned given the training data that was given in a certain sequence that is "remembered" by the model (Durna, 2024).
3. The Hidden layers is updated based on the "remembered" previous hidden stateand the new input (Durna, 2024). 
4. The weights/parameters are shared and used with each input that is in the sequence, this increases the efficiency of the model (Durna, 2024).
5. The models RNN and LSTM 'memorizes" the patterns to be used for latter use in the prediction and identification of the sentiment of reviews that is has not seen before (Shraddha, 2024).

### Cleaning Text data

1. Clearing text from punctuation (regexp_replace)
2. Tokenization (Tokenizer)
3. Delete stop words (StopWordsRemover)
4. Stematization (SnowballStemmer)
5. Filtering short words (udf)

(Profsoft, 2022)

```python
def clean_text(doc):
	# split into tokens by white space
	tokens = re.sub("[^\w]", " ", doc).split()
	# remove punctuation from each token
	table = str.maketrans('', '', string.punctuation)
	tokens = [w.translate(table) for w in tokens]
	# remove remaining tokens that are not alphabetic
	tokens = [word for word in tokens if word.isalpha()]
	# convert back to a string
	tokens = " ".join(tokens)
	return tokens

def stem_lem(doc):
    # split into tokens by white space
	tokens = re.sub("[^\w]", " ", doc).split()
    # Stemming   (GeeksforGeeks 2023)
	porter = PorterStemmer()
	tokens = [porter.stem(word) for word in tokens]
	# Lemmatize tokens
	wordnet_lemmatizer = WordNetLemmatizer()
	tokens = [wordnet_lemmatizer.lemmatize(word) for word in tokens]	
 	# filter out short tokens
	tokens = [word for word in tokens if len(word) > 1]
	# convert back to a string
	tokens = " ".join(tokens)
	return tokens
 

def remove_stopwords(doc):
 	# split into tokens by white space
	tokens = re.sub("[^\w]", " ", doc).split()	
	# filter out short tokens
	tokens = [word for word in tokens if len(word) > 1]
 	#  lowercase letters
	tokens = [word.lower() for word in tokens]
	# filter out stop words
	stop_words = set(stopwords.words('english'))
	tokens = [w for w in tokens if not w in stop_words]
 	# convert back to a string
	tokens = " ".join(tokens)
	return tokens
```

The 'Adam' optimiser is used in the RNN models to optimize the gradient descent, since it does not require a lot of memory and is more efficient (GeeksforGeeks, 2024).

Simple RNN

```python
# Build and train the SimpleRNN model
model = Sequential()
model.add(Embedding(vocab_size,embedding_dim,input_length = max_len))
model.add(GlobalAveragePooling1D())
model.add(Dense( 24,activation='relu'))
model.add(Dropout(drop_value))
model.add(Dense(1,activation ='sigmoid'))

model.compile(loss ='binary_crossentropy',optimizer='adam', metrics=['accuracy'])

num_epochs =30
early_stop = EarlyStopping(monitor='val_loss',patience=2)

history_rnn = model.fit(training_padded,y_train,epochs=num_epochs, validation_data=(testing_padded, y_test),callbacks =[early_stop], verbose =2)
```
